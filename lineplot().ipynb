```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
temperature=[30,29,34,36,45.6,40.7,39.9]
df=pd.DataFrame({'days':days,'temperature':temperature})
sns.lineplot(x='days',y='temperature',data=df)
plt.show()
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.lineplot(x="total_bill",y="tip",data=tips_df)
```

```python
sns.lineplot(x="size",y="tip",data=tips_df)
```

```python
sns.lineplot(x="size",y="tip",data=tips_df,hue="sex")
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(10,5))
sns.lineplot(x="size",y="tip",data=tips_df,hue="sex",style='sex',palette='BuPu',dashes=False,markers=['>',"o"])
plt.title("Line Plot",fontsize=20)
plt.xlabel("Size",fontsize=15)
plt.ylabel("Tips",fontsize=15)
plt.show()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(10,5))
sns.lineplot(x="size",y="tip",data=tips_df,hue="day",style='day',palette='BuPu',dashes=False,markers=['>',"o",'<','^'])
plt.title("Line Plot",fontsize=20)
plt.xlabel("Size",fontsize=15)
plt.ylabel("Tips",fontsize=15)
plt.savefig('D:\\Line Plot seaborn')
plt.show()
```
