```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic_df=sns.load_dataset('titanic')
titanic_df
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df)
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex')
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who')
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who',size='who',sizes=(100,200))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='who',y='fare',data=titanic_df,hue='alive',style='alive',size='who',sizes=(100,200),palette='hot',alpha=0.9)
```
